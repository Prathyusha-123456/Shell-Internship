```python
# Create Flood_Risk column based on ANNUAL rainfall
annual_mean = df['ANNUAL'].mean()
annual_std = df['ANNUAL'].std()
threshold = annual_mean + 1.5 * annual_std

df['Flood_Risk'] = (df['ANNUAL'] > threshold).astype(int)

print("Annual mean rainfall:", annual_mean)
print("Flood threshold:", threshold)
print("Flood Risk distribution:\n", df['Flood_Risk'].value_counts())

print("\nColumns now:", df.columns.tolist())
```

 FEATURE ENGINEERING


```python
df['Decade'] = (df['YEAR'] // 10) * 10   # decade grouping (19011910, etc.)
df['Year_mod10'] = df['YEAR'] % 10       # position within decade
```

```python

# 1. Seasonal totals
df['Monsoon_Rain'] = df[['JUN','JUL','AUG','SEP']].sum(axis=1)
df['PreMonsoon_Rain'] = df[['MAR','APR','MAY']].sum(axis=1)
df['PostMonsoon_Rain'] = df[['OCT','NOV','DEC']].sum(axis=1)

# 2. Rolling features (3-year moving average & anomaly)
df['Annual_MA3'] = df['ANNUAL'].rolling(window=3, min_periods=1).mean()
df['Rainfall_Anomaly'] = df['ANNUAL'] - df['Annual_MA3']

df.head()
```

```python
annual_mean = df['ANNUAL'].mean()
df['Rainfall_Anomaly'] = df['ANNUAL'] - annual_mean
```
